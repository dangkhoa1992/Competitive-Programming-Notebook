## [Maximum Subarray Sum](https://leetcode.com/problems/maximum-subarray/)
- Given an integer array nums, find the contiguous subarray (containing at least one number) which has the largest sum and return its sum.

- Example:

```
Input: [-2,1,-3,4,-1,2,1,-5,4],
Output: 6
Explanation: [4,-1,2,1] has the largest sum = 6.
```

#### Divide and Conquer O(nlog(n))
- max subarraySum from x

```C++
class Solution {
public:
    vector<int> _A;

    int __max3(int x1, int x2, int x3) {
        return max(x1, max(x2, x3));
    }
    int __get_max_cross_sum(int x, int l, int r) {
        // get max sum from left side
        int ptr = 0;
        int max_sum_left = INT_MIN;
        for(int i=x; i>=l; --i) {
            ptr += _A[i];
            max_sum_left = max(max_sum_left, ptr);
        }

        // get max sum from right side
        ptr = 0;
        int max_sum_right = INT_MIN;
        for(int j=x+1; j<=r; ++j) {
            ptr += _A[j];
            max_sum_right = max(max_sum_right, ptr);
        }

        // max sum cross = max_sum_left + max_sum_right
        return max_sum_left + max_sum_right;
    }

    int get(int l, int r) {
        if(l == r) return _A[l];

        // Choose x = mid(l,r)
        int x = l + (r-l)/2;

        int max_sum_left = get(l,x);
        int max_sum_right = get(x+1,r);
        int max_cross_sum = __get_max_cross_sum(x, l, r);

        return __max3(max_sum_left, max_sum_right, max_cross_sum);
    }
    int maxSubArray(vector<int>& nums) {
        _A = nums;
        return get(0, _A.size()-1);
    }
};
```
