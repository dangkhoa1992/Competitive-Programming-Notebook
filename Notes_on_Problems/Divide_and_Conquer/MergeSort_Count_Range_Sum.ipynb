## [Count of Range Sum](https://leetcode.com/problems/count-of-range-sum/)
- Given an integer array nums, return the number of range sums that lie in [lower, upper] inclusive.
- Range sum S(i, j) is defined as the sum of the elements in nums between indices i and j (i â‰¤ j), inclusive.
- **Example**

```
Input: nums = [-2,5,-1], lower = -2, upper = 2,
Output: 3 
Explanation: 
    The three ranges are : [0,0], [2,2], [0,2] and their respective sums are: -2, -1, 2.
```

#### Solution

```C++
class Solution {
public:
    int merge(vector<long> &A, vector<long> &temp, int l, int m, int r, int lower, int upper) {
        // count sum ranges: 2 pointers
        int cnt = 0;
        int a, b;
        a = b = m+1;
        for(int i=l; i<=m; ++i) {
            while(a <= r && A[a] - A[i] < lower) ++a;
            while(b <= r && A[b] - A[i] <= upper) ++b;
            cnt += b - a;
        }

        // merge sort
        int i = l;
        int j = m + 1;
        int cur = 0;
        while(i <= m || j <= r) {
            if(i > m) temp[cur++] = A[j++];
            else if(j > r) temp[cur++] = A[i++];
            else if(A[i] <= A[j]) temp[cur++] = A[i++];
            else temp[cur++] = A[j++];
        }
        for(int i=0; i < cur; ++i) A[l + i] = temp[i];

        return cnt;
    }
    int mergeSort(vector<long> &A, vector<long> &temp, int l, int r, int lower, int upper) {
        if(l >= r) return 0;

        int m = l+(r-l)/2;
        return  mergeSort(A, temp, l, m, lower, upper) + \
                mergeSort(A, temp, m+1, r, lower, upper) + \
                merge(A, temp, l, m, r, lower, upper);
    }
    int countRangeSum(vector<int>& nums, int lower, int upper) {
        int N = nums.size();
        vector<long> temp(N+1);

        // Cache prefix sums
        vector<long> prefix_sums(N+1, 0);
        for(int i=0; i<N; ++i) prefix_sums[i+1] = prefix_sums[i] + nums[i];

        return mergeSort(prefix_sums, temp, 0, N, lower, upper);
    }
};
```
