
## All Permutations of a string 1
- Find n! permutation of a string
- Example s = "ABC":

```
ABC
ACB
BAC
BCA
CBA
CAB
```


#### Solution
- dfs the tree

<img src="./img/5.jpg" alt="drawing" width="650"/>


```C++
string s = "ABC";
int n = s.size();

// f(0)
vector<bool> chosen(n,false);
string permu;
void f(int x) {
    if(x == n) {
        cout << permu << endl;
        return;
    }

    for(int c=0; c<n; ++c) {
        if(chosen[c]) continue;
        
        permu.push_back(s[c]);
        chosen[c] = true;

        f(x+1);

        permu.pop_back();
        chosen[c] = false;
    }
}
```

````python
def generate_permute(perm, n):
    if len(perm) == n:
        print(perm)
        return

    for k in range(n):
        if k not in perm:
            generate_permute(perm + [k], n)

generate_permute(perm=[], n=4)
````

## All Permutations of a string 2
- Find n! permutation of a string
- Example s = "ABC":

```
ABC
ACB
BAC
BCA
CBA
CAB
```

#### Solution
- dfs the tree
<img src="./img/1.jpg" alt="drawing" width="800"/>



```C++
string s = "ABC";
int n = s.size();

// f(0)
void f(int x) {
    if(x == n) {
        cout << s << endl;
        return;
    }

    for(int c=x; c<n; ++c) {
        swap(s[x], s[c]);

        f(x+1);
        
        // Backtrack
        swap(s[x], s[c]);
    }
}
```
