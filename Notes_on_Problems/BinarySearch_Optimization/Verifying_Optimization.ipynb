##  [Split Array Largest Sum](https://leetcode.com/problems/split-array-largest-sum/)
- Split an array into 
    + m non-empty continuous subarrays
    + minimize the largest sum among these m subarrays.
- Example

```
[7,2,5,10,8], m = 2

4 ways of spliting
    [7] [2,5,10,8] largest_sum = 25
    [7,2] [5,10,8] largest_sum = 23
    [7,2,5] [10,8] largest_sum = 18
    [7,2,5,10] [8] largest_sum = 24
    
=> minimum largest_sum = 18
```

#### Solution - O(Nlog(N))
- Range of max_val:

``` 
    0        [step=1]                         total_sum       
    |--|--|--|--|--|--|--|--|--|--|--|--|--|--|--|
```
- m subarray -> m-1 cuts
- for each max_val we can check if the array is splitable with m-1 number of cuts and the largest segment <= max_val-> Find the minimum max_val that splitable 

```
    [      ][  ][ ]
```

```C++
class Solution {
public:
    int N;
    /*--- Check if nums can be split by max_val and cuts O(n)---*/
    bool canSplit(const vector<int>& nums, int cuts, int max_val) {
        int sum_ = 0;
        for(auto &num : nums) {
            if(sum_ + num <= max_val) sum_ += num;
            else {
                --cuts;
                if(cuts < 0) return false;
                sum_ = num;
            }
        }
        return true;
    }
    int splitArray(vector<int>& nums, int M) {
        N = nums.size();

        int left=0, right = 0, mid;
        for(auto &num : nums) right += num;
        
        // Binary search - Find the minimum max_val possible
        while(left <= right) {
            mid = left + (right-left)/2;
            if(canSplit(nums, M-1, mid))
                right = mid-1;
            else
                left = mid+1;
        }
        return left;
    }
};
```
