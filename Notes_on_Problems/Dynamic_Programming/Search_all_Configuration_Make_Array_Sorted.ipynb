## Make an Array sorted
- Give an array A[1:N]
- We can add 1 or -1 to each element of A[] to become A'[] sorted in increasing order $(A'[1] \leq A'[2] \leq A'[3] \leq ... \leq A'[N])$
- Find the minimum cost to make an Array A -> Sorted Array A'

#### Example

```
Input
3
1 2 1

Output
1

Explanation
[1,2,1] -> [1,1,1]


Input
2
5 10

Output
0


Input
7
1 4 2 3 1 4 4

Output
4

Explanation
[1,4,2,3,1,4,4] -> [1,2,2,3,3,4,4]
```



## Solution $O(n^3)$
#### Search space
- A[i] can be replaced by A'[i]
    + $min\_val \leq A'[i] \leq max\_val$
    + $A'[i] \leq A'[i+1]$
- We search all possible configuration that produce the total min cost

- Relation
$$f(i,y) = min_{min\_val \leq x \leq y} (f(i-1,x) + cost(i,y))$$
- In which
    + y = A'[i]
    + x = A'[i-1],  $min\_val \leq x \leq y$
    + cost(i,y) = $|y - A[i]|$

#### Code

```C++
int max_val;
int min_val;

int dp[2003][1003];
int get(int i, int y) {
    if(i<1 || y<min_val) return inf;
    if(i==1) return abs(y-A[1]);
    if(dp[i][y] != inf) return dp[i][y];

    FOR(x,min_val,y+1) 
        dp[i][y] = min(dp[i][y], get(i-1,x) + abs(y-A[i]));

    return dp[i][y];
}
int solveDP() {
    // Find the range
    int max_val = 0;
    int min_val = inf;
    FOR(i,1,N+1) {
        max_val = max(max_val, A[i]);
        min_val = min(min_val, A[i]);
    } 

    // DP 
    int ans = inf;
    ms(dp, 0x7f);
    FOR(v,min_val,max_val+1)
        ans = min(ans, get(N,v));
    return ans;
}
```

## Optimized Solution $O(n^2)$
