## Problem
- You’re planning a company party. You’d like to invite the coolest people, and you’ve assigned each
one of them a fun factor — the more the fun factor, the cooler is the person. 
    + You want to maximize the total fun factor (sum of the fun factors of all the invited people)
    + However, you can’t invite everyone, because if the direct boss of some invited person is also invited, it will be awkward. Find out what is the maximum possible total fun factor.

#### Tree
```C++
int fun[MAXV];
vector<int> e[MAXV];
    // root = 1
```

## Solution
- DFS tree
- DP(v, c):
    + v: Optimal solution at node v
    + c: select(0) or skip(1) v
- Cases
    + Case 1: v is selected - Skip all children
    + Case 2: v is skipped - Select a subset of children

```C++
int dp[MAXV][2];
    // dp[v][0]: not skip v
    // dp[v][1]: skip v
int state[MAXV];
    // 0: not visited
    // 1: pre visited
    // 2: post visited
void dfs(int v) {
    // dfs
    state[v] = 1;

    FOR(i,0,sz(e[v])) {
        if(state[e[v][i]] == 0)
            dfs(e[v][i]);
    }

    state[v] = 2;

    // Not Skip v - Skip all children
    int temp = 0;
    FOR(i,0,sz(e[v])) {
        // v children = post order processed node
        if(state[e[v][i]] != 2)
            continue;
        temp += dp[e[v][i]][1];
    }
    dp[v][0] = max(dp[v][0], temp + fun[v]);

    // Skip v - select a subset of children
    temp = 0;
    FOR(i,0,sz(e[v])) {
        // v children = post order processed node
        if(state[e[v][i]] != 2)
            continue;
        temp += max(dp[e[v][i]][0], dp[e[v][i]][1]); // select or skip child
    }
    dp[v][1] = max(dp[v][1], temp);
}

int solveDP() {
    // init
    ms(state, 0);
    ms(dp,-1);
    
    // Base cases
    FOR(v,1,V+1) {
        dp[v][0] = fun[v];
        dp[v][1] = 0;
    }

    // Solve DP
    dfs(1);
    return max(dp[1][0], dp[1][1]);
}
```
