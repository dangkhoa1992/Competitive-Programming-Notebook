## Bit counting
- Count number of the bit strings
    + size = n
    + without 2 consecutive bit '1'
- Example n = 4
<img src="./img/6.jpg" alt="drawing" width="600"/>


#### Top down
```C++
// i = i_th digit, x = end_bit (0 or 1)
int dp[103][2];
int get(int i, int x) {
    if(i == 0) return 0;
    if(i == 1) return 1;
    if(dp[i][x] != -1) return dp[i][x];

    if(x == 0) dp[i][x] = get(i-1, 0) + get(i-1, 1);
    if(x == 1) dp[i][x] = get(i-1, 0);

    return dp[i][x];
}
int solveDP(int n) {
    ms(dp, -1);
    return get(n,0) + get(n,1);
}
```


#### Bottom Up

```C++
/* --------------- Solution 1 - O(n)-------------- */
int dp[103][2];
int sol1(int n) {
    // Base case
    dp[1][0] = 1;
    dp[1][1] = 1;

    // DP
    FOR(i,1,n+1) {
        dp[i+1][0] = dp[i][0] + dp[i][1];
        dp[i+1][1] = dp[i][0];
    }
    return dp[n][0] + dp[n][1];
}
```

```C++
/* ------------- Solution 2 - O(log(n))---------- */
int sol2(int n) {
    vvi X = { {1, 1},
              {1, 0} };
    vvi dp_1 = { {1},
                 {1} };

    vvi res = matrixMul(matrixPow(X, n-1), dp_1); 
    return res[0][0] + res[1][0];
}
```

## [Path counting](http://codeforces.com/problemset/problem/166/E)

<img src="./img/5.jpg" alt="drawing" width="900"/>

#### Top down
```C++
int dp[1003][2];
            // 0 = A,B,C
            // 1 = D
int get(int i, int x) {
    if(i == 1 && x == 0) return 1;
    if(i == 1 && x == 1) return 0;
    if(dp[i][x] != -1) return dp[i][x];

    if(x == 0) dp[i][x] = 2*get(i-1, 0) + get(i-1, 1);
    if(x == 1) dp[i][x] = 3*get(i-1, 0);

    return dp[i][x];
}
int solveDP(int n) {
    ms(dp, -1);
    return get(n, 1);
}
```


#### Bottom Up

```C++
/* --------------- Solution 1 - O(n)-------------- */
int dp[1003][2];
int sol1(int n) {
    // 0 = A,B,C
    // 1 = D
    dp[1][0] = 1; 
    dp[1][1] = 0;
    FOR(i,1,n+1) {
        dp[i+1][0] = 2*dp[i][0] + dp[i][1];
        dp[i+1][1] = 3*dp[i][0];
    }
    return dp[n][1];
}
```
```C++
/* ------------- Solution 2 - O(log(n))------------ */

int sol2(int n) {
    vvi X = { {2, 1},
              {3, 0} };
    vvi dp_0 = { {1},
                 {0} };
    return matrixMul(matrixPow(X, n-1), dp_0)[1][0];
}
```
