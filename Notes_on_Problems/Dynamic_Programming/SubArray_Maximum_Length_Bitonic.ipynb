#### Bitonic subarray
-  A subarray A[i, j] is bitonic if
    + contain a purely increasing subsequence + a purely decreasing subsequence
    + Exist a k
        + A[x] < A[x+1], x in range [i, k]
        + A[x] > A[x+1], x in range [k, j]
- Example:
    + 15 20 6 4 2, peak = 20
    + 1 2 3 4 2, peak = 4
    + 10 20 30 40, peak = 40
    + 40 30 20 10, peak = 40

# [Problem](https://www.geeksforgeeks.org/maximum-length-bitonic-subarray/)
- Given an array A
- Find the maximum length of a subarray that are bitonic

#### Example


```
Input:
6
12 4 78 90 45 23

Output: 5
Explanation: bitonic subarray is [4, 78, 90, 45, 23]


Input:
8
20 4 1 2 3 4 2 10

Output: 5
Explanation: bitonic subarray is [1, 2, 3, 4, 2]


Input:
1
10

Output: 1


Input:
4
10 20 30 40

Output: 4


Input:
4
40 30 20 10

Output: 4
```

#### Solution O(n)

```cpp
int N;
vector<int> A;

int dp_inc[100005];
int dp_dec[100005];
int sol() {
    // Init
    for(int i = 0; i < N; ++i) {
        dp_inc[i] = 1;
        dp_dec[i] = 1;
    }

    // Cache: Increasing sequence array: left -> right
    for(int i = 0; i < N-1; ++i) {
        if(A[i] < A[i+1]) dp_inc[i+1] = dp_inc[i] + 1;
    }

    // Cache: Decreasing sequence array: left <- right
    for(int i = N-1; i >= 1; --i) {
        if(A[i-1] > A[i]) dp_dec[i-1] = dp_dec[i] + 1;
    }

    // Solve
    int ans = 0;
    for(int i = 0; i < N; ++i) {
        ans = max(ans, dp_inc[i] + dp_dec[i] - 1);
    }
    return ans;
}
```
