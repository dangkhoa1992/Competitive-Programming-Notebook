# 2. Probability

## 2.1 Independent vs Dependent

#### Independent Events
- Independent Events
    + 2 events are independent when the outcome of the first event **does not influence** the outcome of the second event
    - Example: 
        + Getting a 6 when rolling the first die does not affect the outcome of rolling the second die

```python
P(X and Y) = P(X) * P(Y)
```

- **Problem**: The probability of getting three 6s if we roll 3 dices
- **Solution**
    + The probability of getting a 6 on one die: $P("6") = \frac{1}{6}$
    + The probability of getting three 6s on 3 dies: $P("666") = \frac{1}{6}*\frac{1}{6}*\frac{1}{6} = \frac{1}{216}$


#### Dependent Events
- Dependent Events
    + 2 events are dependent when the outcome of the first event **affects** the outcome of the second event
    + Example:
        + Choose 2 redcard from the same deck

```python
P(X and Y) = P(X) * P(Y after X has occurred)
           = P(X) * P(Y|X)
```

- **Problem**: The probability of getting 2 redcards from the same deck
- **Solution**
    + Draw 1st red from 26 reds, 26 blacks: $P(\text{1st red}) = \frac{26}{52} = \frac{1}{2}$
    + Draw 2nd red after 1st red: 25 reds, 26 blacks: $P(\text{2nd red after 1st red}) = \frac{25}{51}$
    + $P(\text{2 red from same deck}) = \frac{1}{2} * \frac{25}{51} = \frac{25}{102}$

## 2.2 Exclusive vs Inclusive

#### Mutually Exclusive Events
- Mutually Exclusive Events
    + 2 events are mutually exclusive when it is **impossible to happen together** at the same time
    + Examples
        + Draw a card from a deck
            + Event X: the card is a King
            + Event Y: the card is a Queen  
                => X and Y are mutually exclusive
        + Toss a coin
            + Event X: get a head
            + Event Y: get a tail  
                => X and Y are mutually exclusive

```python
P(X and Y) = 0
P(X or Y) = P(X) + P(Y)
```

- **Problem**: Draw a card from a deck
    + Probability of this card is both King and Queen
    + Probability of this card is King or Queen
- **Solution**
    + Probability of this card is both King and Queen: $P(\text{King and Queen}) = 0$
    + Probability of this card is King or Queen: 
        + $P(\text{King or Queen}) = P(King) + P(Queen) = \frac{1}{13} + \frac{1}{13} = \frac{2}{13}$

#### Inclusive Events
- Inclusive events = events that **can happen at the same time**
    + Example: Draw a card from a deck
        + Event X: the card is a Queen
        + Event Y: the card is a Heart 
            => X and Y are inclusive

<img src="./img/8.jpg" width="190"/>

```python

P(X or Y) = P(X) + P(Y) âˆ’ P(X and Y)
P(X and Y) = P(X) * P(Y|X) = P(Y) * P(X|Y) # Dependent events
P(X and Y) = P(X) * P(Y)                   # Independent events
```

- **Problem** : Draw a card from a deck, probability that the card is a Queen or a Heart
- **Solution 1**:
$$\begin{split}
    P(\text{Queen or Heart}) &= P(Queen) + P(Heart) - P(\text{Queen and Heart}) \\
        &= \frac{4}{52} + \frac{13}{52} - \frac{1}{52} = \frac{16}{52} = \frac{4}{13}
\end{split}$$

- **Solution 2**: Heart and Queen are mutally dependent events

$$\begin{split}
    P(\text{Queen or Heart}) &= P(Queen) + P(Heart) - P(\text{Queen and Heart}) \\
    &= P(Queen) + P(Heart) - P(\text{Queen}) \cdot P(\text{Heart | Queen}) \\
    &= \frac{4}{52} + \frac{13}{52} - \frac{4}{52}*\frac{1}{4} = \frac{16}{52} = \frac{4}{13}
\end{split}$$

# 3. Expectation

$$ E[X] = \sum\limits_ix_iP(x_i) = x_1P(x_1) + x_2P(x_2) + \dots + x_nP(x_n) $$

- where
    + $[x_1, x_2, \dots, x_n]$: all possible outcomes
    + $p_1 + p_2 + \dots + p_n = 1$

- **Problem**: Tossing a fair coin 3 times. Let X denote the number of heads which appear, calc E(X)
- **Solution**:
    + All possible outcomes: 0 heads, 1 head, 2 heads, 3 heads

$$\begin{split}
E(X) &= 0*p(0) + 1*p(1) + 2*p(2) + 3*p(3) \\
     &= 0*\frac{1}{8} + 1*\frac{3}{8} + 2*\frac{3}{8} + 3*\frac{1}{8} \\
     &= 1.5
\end{split}$$

- Properties
$$\begin{split}
& E[X + Y] = E[X] + E[Y] \\
& E[X * Y] = E[X] * E[Y]
\end{split}$$
